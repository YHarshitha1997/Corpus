### Removing stop words and punctuations in text files

```python
from nltk.tokenize import word_tokenize
#import string
#from nltk.corpus import stopwords
lst_txt = []

for i in textfileslist:
    with open('/Users/harshithayentrapragada/Desktop/601/WEEK-9/week_10_txt_and_docx/{}'.format(i),'rt') as f:
        text = f.read()
        f.close()
        tokens = word_tokenize(text)
        tokens = [w.lower() for w in tokens]
        table = str.maketrans('', '', string.punctuation)
        stripped = [w.translate(table) for w in tokens]
        words1 = [word for word in stripped if word.isalpha()]
    
        stop_words = set(stopwords.words('english'))
        words1 = [w for w in words1 if not w in stop_words]
        print(words1)
        lst_txt.append(words1)
```

### Merging two docx in one docx file

```python
# https://stackoverflow.com/questions/24872527/combine-word-document-using-python-docx
    
from docx import Document

files = ['week_10_document1.docx', 'week_10_document2.docx']

def combine_word_documents(files):
    merged_document = Document()

    for index, file in enumerate(files):
        sub_doc = Document(file)

        # Don't add a page break if you've reached the last file.
        if index < len(files)-1:
           sub_doc.add_page_break()

        for element in sub_doc.element.body:
            merged_document.element.body.append(element)

    merged_document.save('merged.docx')

combine_word_documents(files)
```

```python
my_text = docx2txt.process("merged.docx")
print(my_text)
```

### removing punctuations and stop words

```python
lst_docx = []
tokens = word_tokenize(my_text)
tokens = [w.lower() for w in tokens]
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens]
words = [word for word in stripped if word.isalpha()]
    
stop_words = set(stopwords.words('english'))
words = [w for w in words if not w in stop_words]
print(words)
lst_docx.append(words)
```
